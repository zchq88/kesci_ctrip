# 尝试增加2014比例浮动

```python
product_quantity=pd.read_csv('product_quantity.txt')
product_quantity['product_month']=product_quantity['product_date'].apply(lambda x: x[:7])
product_quantity=product_quantity[product_quantity.product_month<='2015-00']
train_month=product_quantity.groupby(['product_id','product_month']).sum()['ciiquantity'].unstack()
train_month
```

```python
train_month['avg']=train_month.T.sum()/12
#train_month=pd.merge(train_month,average_all,on='product_id',how='left').fillna(0)
train_month['01']=train_month['2014-01']/train_month['avg']
train_month['02']=train_month['2014-02']/train_month['avg']
train_month['03']=train_month['2014-03']/train_month['avg']
train_month['04']=train_month['2014-04']/train_month['avg']
train_month['05']=train_month['2014-05']/train_month['avg']
train_month['06']=train_month['2014-06']/train_month['avg']
train_month['07']=train_month['2014-07']/train_month['avg']
train_month['08']=train_month['2014-08']/train_month['avg']
train_month['09']=train_month['2014-09']/train_month['avg']
train_month['10']=train_month['2014-10']/train_month['avg']
train_month['11']=train_month['2014-11']/train_month['avg']
train_month['12']=train_month['2014-12']/train_month['avg']
train_month=train_month.ix[:,['01','02','03','04','05','06','07','08','09','10','11','12']]
```

# 数据填写

```python
import sys

def _map(data,exp):
    #data.loc['ciiquantity_month',1]=100
    for index, row in data.iterrows():   # 获取每行的index、row
        data.ix[index,'ciiquantity_month']=exp(row['product_id'],row['product_month'],row['average_all'])
        #print(row['ciiquantity_month'])
        sys.stdout.write('\r{}'.format(index))
    return data

def set_data(_id,_mm,_avg):
    _x=1
    try:
        _x=train_month.ix[_id,_mm[5:7]]
    except:      
        _x=1
    #print(_id,_mm[5:7],_avg,_x,_avg*_x)    
    return _avg*_x

submission=evaluation

submission.shape

col=['product_id','product_month','ciiquantity_month']

submission.columns=col

out=pd.merge(submission,average_all,on='product_id',how='left').fillna(153)

out.apply(lambda x: sum(x.isnull()))

#out=_map(out, set_data)

out.ciiquantity_month=out.average_all.apply(lambda x: x)

out.drop(['average_all'],axis=1,inplace=True)

out.to_csv('average9months_153.txt',index=False)
out
```
